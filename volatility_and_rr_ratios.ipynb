```python
stocklist=['TITAN', "BPCL", "LT", "TATASTEEL", "SBILIFE", "HDFCLIFE", "RELIANCE", "BHARTIARTL", "SUNPHARMA", "NTPC", "POWERGRID", "INDUSINDBK", "HEROMOTOCO",
            "HINDUNILVR", "COALINDIA", "CIPLA", "ULTRACEMCO", "HDFCBANK", "JSWSTEEL", "HINDALCO", "NESTLEIND", "TATACONSUM", "BRITANNIA", "HDFC", 
            "ONGC", "DRREDDY", 'ITC', "DIVISLAB", 'MARUTI', "BAJAJ-AUTO", 'TCS', "GRASIM", 'ASIANPAINT', 'INFY', 'APOLLOHOSP', 'ICICIBANK', 'WIPRO', 
            'TATAMOTORS', 'EICHERMOT', 'KOTAKBANK', 'M&M', 'SBIN', 'BAJAJFINSV', 'AXISBANK', 'HCLTECH', 'BAJFINANCE', 'UPL', 'TECHM', 'ADANIPORTS', 
            'ADANIENT']
stocks =['^NSEI']+[i+'.NS' for i in stocklist]
print((stocks))
```

IMPORTING DATA FROM YAHOO FINANCE

```python
end = dt.datetime(2023,2,1)
start = dt.datetime(2022,2,1)
yf.pdr_override()
df=pdr.get_data_yahoo(stocks,start,end)
close=df.Close
close.head()
```

CALCULATING DAILY LOG RETURNS

```python
log_returns = np.log(df.Close/df.Close.shift(1)).dropna()
log_returns
```

```python
daily_std=log_returns.std()
daily_std
```

```python
#calculating annual volatility using the number of trading days in a year
annualized_vol = daily_std *np.sqrt(252)
annualized_vol*100
```

```python
trading_days = 60
volatility = log_returns.rolling(window=trading_days).std()*np.sqrt(trading_days)
```

PLOTTING VOLATILITY

```python
pyo.plot(volatility,kind='line').update_layout(autosize=False, width = 1200, height = 300)
```

CALCULATING SHARPE RATIO

Sharpe ratio divides a portfolio's excess returns by a measure of its volatility to access risk adjusted performance.

A rolling sharpe ratio is calculated to get a clear understanding of the performance and also to help in plotting.

```python
rf = 0.01/252
sharpe_ratio = (log_returns.rolling(window=trading_days).mean()-rf)*trading_days/volatility
```

```python
pyo.plot(sharpe_ratio,kind='line').update_layout(autosize=False, width = 1200, height = 300)
```

SORTINO RATIO 

It helps in differentiating between harmful volatility and total volatility by using the asset's standard deviation of negative portfolio returns.

```python
sortino_vol = log_returns[log_returns<0].rolling(window=trading_days,center=True,min_periods=10).std()*np.sqrt(trading_days)
sortino_ratio = ((log_returns.rolling(window=trading_days).mean()-rf)*trading_days/sortino_vol).dropna()
```

```python
pyo.plot(sortino_ratio,kind='line').update_layout(autosize=False, width = 1200, height = 300)
```
